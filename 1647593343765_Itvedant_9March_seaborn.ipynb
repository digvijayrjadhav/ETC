## standard deviation

```python
#standard deviation
#it is a square root of avg of squared deviations from mean
#std=sqrt(mean(abs(x-xmean())**2))
import numpy as np
```

```python
a=np.array([1,2,3,4])
a
```

```python
#mean of array=xmean = 2.5
#x=1,xmean=2.5 x-xmean=1-2.5=-1.5,abs(-1.5)=1.5^2=2.25
#x=2,xmean=2.5,x-xmean=2-2.5=-0.5,abs(-0.5)=0.5^2=0.25
#3-2.5=0.5^=0.25
#4-2.5=1.5^2=2.25
#sqared deviations are [2.25,0.25,0.25,2.25]
#mean(sqaured deviations) =5/4=1.25
#sqrt(1.25)=1.1180
#std=1.1180
np.std(a)
```

```python
#variance is the measure of how far a value in the dataset lies from mean value.
#it is measured using standard deviation
```

```python
#variance = avarage of squared deviations
np.var(a)
```

### merging

### pd.merge(left,right,on="comman column",how='inner')
#left=first df
#right=second df
#on = consider comman column in 2 dataframes and merge on that column

```python
import pandas as pd
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
registrations
```

```python
logins
```

```python
df=pd.merge(registrations,logins,on='name') #how='inner' is bydefault
df
```

```python
pd.merge(registrations,logins,on="name",how='inner')
#how='inner' means it will take comman records in both tables 
```

```python
pd.merge(registrations,logins,on="name",how='left')
#how=left ,it will select all records from left(dataframe1) and comman records from right(dataframe2)
```

```python
pd.merge(registrations,logins,on="name",how='right')
#how=right ,it will select all records from right(dataframe2) and comman records from left(dataframe1)
```

```python
pd.merge(registrations,logins,on="name",how='outer')
#how= outer ,it will select all records from both tables
```

```python
registrations = registrations.set_index("name")
registrations
```

```python
logins
```

```python
#no comman column in both dataframes
pd.merge(registrations,logins,left_index=True,right_on='name')
```

```python
registrations = registrations.reset_index()
registrations
```

```python
logins
```

```python
registrations.columns=['reg_name','reg_id']
registrations
#column name is changed
```

```python
pd.merge(registrations,logins,left_on='reg_name',right_on='name')
```

```python
registrations.columns=['name','id']
registrations
```

```python
logins.columns=['id','name']
logins
```

```python
pd.merge(registrations,logins,on='name')
```

```python
pd.merge(registrations,logins,on='name',suffixes=('_reg','_log'))
```
